## Strategy implementation (macd)

### Subtask:

**Reasoning**:
Define the trading signals based on the MACD and signal line crossover and add the 'signal' column to the DataFrame.



```python
# Create a new column 'signal' initialized to 0 (no signal)
df['signal'] = 0

# Generate buy signals: when MACD crosses above the Signal Line
# This occurs when MACD was below Signal Line in the previous period and is above or equal in the current period
df.loc[(df['macd'].shift(1) < df['signal_line'].shift(1)) & (df['macd'] >= df['signal_line']), 'signal'] = 1

# Generate sell signals: when MACD crosses below the Signal Line
# This occurs when MACD was above Signal Line in the previous period and is below or equal in the current period
df.loc[(df['macd'].shift(1) > df['signal_line'].shift(1)) & (df['macd'] <= df['signal_line']), 'signal'] = -1

# Display the first few rows to show the added signal column
print("\nDataFrame with Trading Signals:")
display(df.head())

# Display the tail of the DataFrame to see potential recent signals
print("\nTail of DataFrame with Trading Signals:")
display(df.tail())

# Display some rows where signals are generated to verify
print("\nRows with Buy Signals (signal == 1):")
display(df[df['signal'] == 1].head())

print("\nRows with Sell Signals (signal == -1):")
display(df[df['signal'] == -1].head())
```

## Backtesting

### Subtask:
Implement a backtesting framework to simulate trades based on the MACD strategy signals using the historical data.


**Reasoning**:
Initialize backtesting variables and iterate through the DataFrame to simulate trades based on the MACD signals, updating the balance and recording trades.



```python
# Initialize backtesting variables
initial_balance = 10000
balance = initial_balance
position = 'none' # 'none', 'long', 'short'
entry_price = 0
trades = []

# Iterate through the DataFrame starting after the initial NaNs from MACD calculation
# Assuming the first non-NaN index is the start of valid signals
start_index = df['signal'].first_valid_index()
if start_index is None:
    print("No valid signals generated. Cannot perform backtesting.")
else:
    print(f"Starting backtest from index: {start_index}")
    for i in range(df.index.get_loc(start_index), len(df)):
        current_date = df.index[i]
        current_price = df['Close'].iloc[i]
        current_signal = df['signal'].iloc[i]

        # Trading logic based on signals
        if current_signal == 1 and position == 'none':
            # Buy signal and no open position - open long
            position = 'long'
            entry_price = current_price
            print(f"{current_date}: BUY signal at {entry_price:.5f}")

        elif current_signal == -1 and position == 'none':
            # Sell signal and no open position - open short
            position = 'short'
            entry_price = current_price
            print(f"{current_date}: SELL signal at {entry_price:.5f}")

        elif current_signal == -1 and position == 'long':
            # Sell signal and long position open - close long
            pnl = (current_price - entry_price)
            balance += pnl
            trades.append({
                'entry_date': df.index[i-1], # Approximate entry date
                'exit_date': current_date,
                'entry_price': entry_price,
                'exit_price': current_price,
                'position': 'long',
                'pnl': pnl
            })
            position = 'none'
            entry_price = 0
            print(f"{current_date}: Close LONG at {current_price:.5f}, PnL: {pnl:.5f}, New Balance: {balance:.2f}")

        elif current_signal == 1 and position == 'short':
            # Buy signal and short position open - close short
            pnl = (entry_price - current_price)
            balance += pnl
            trades.append({
                'entry_date': df.index[i-1], # Approximate entry date
                'exit_date': current_date,
                'entry_price': entry_price,
                'exit_price': current_price,
                'position': 'short',
                'pnl': pnl
            })
            position = 'none'
            entry_price = 0
            print(f"{current_date}: Close SHORT at {current_price:.5f}, PnL: {pnl:.5f}, New Balance: {balance:.2f}")

    # Calculate final results
    total_profit_loss = balance - initial_balance
    print("\n--- Backtesting Results ---")
    print(f"Initial Balance: ${initial_balance:.2f}")
    print(f"Final Balance: ${balance:.2f}")
    print(f"Total Profit/Loss: ${total_profit_loss:.2f}")
    print(f"Total Trades: {len(trades)}")

# The backtesting framework is implemented, but the results are not yet summarized or visualized as per the overall task.
# However, the current subtask of implementing the backtesting framework is complete.
```

```python
# Iterate and Backtest with Optimization for Combined Strategy

# List to store backtesting results for each parameter combination
results_combined_opt = []

# Get the EURUSD DataFrame
if 'EURUSD' not in dataframes:
    print("EURUSD data not found. Cannot perform combined strategy optimization.")
else:
    eurusd_df_combined_opt = dataframes['EURUSD'].copy()
    eurusd_df_combined_opt['Date'] = pd.to_datetime(eurusd_df_combined_opt['Date'])
    eurusd_df_combined_opt.set_index('Date', inplace=True)

    # Loop through all combinations of parameters
    for fast_period in fast_periods_opt:
        for slow_period in slow_periods_opt:
            if slow_period <= fast_period: # Ensure slow_period > fast_period for MACD
                continue
            for signal_period in signal_periods_opt:
                for rsi_timeperiod in rsi_timeperiods_opt:
                    for sl_pct in stop_loss_pct_range:
                        for tp_pct in take_profit_pct_range:

                            # Ensure TP is greater than SL for meaningful profit
                            if tp_pct <= sl_pct:
                                continue

                            print(f"Backtesting with params: MACD({fast_period},{slow_period},{signal_period}), RSI({rsi_timeperiod}), SL={sl_pct:.2%}, TP={tp_pct:.2%}")

                            # Create a copy of the DataFrame for each iteration
                            df_opt = eurusd_df_combined_opt.copy()

                            # Calculate MACD indicators
                            try:
                                df_opt['macd'], df_opt['signal_line'], df_opt['macd_hist'] = talib.MACD(df_opt['Close'],
                                                                                                    fastperiod=fast_period,
                                                                                                    slowperiod=slow_period,
                                                                                                    signalperiod=signal_period)
                            except Exception as e:
                                print(f"Error calculating MACD for params ({fast_period}, {slow_period}, {signal_period}): {e}")
                                continue # Skip this combination

                            # Calculate RSI indicator
                            try:
                                df_opt['rsi'] = talib.RSI(df_opt['Close'], timeperiod=rsi_timeperiod)
                            except Exception as e:
                                print(f"Error calculating RSI for timeperiod ({rsi_timeperiod}): {e}")
                                continue # Skip this combination

                            # Define combined strategy signals
                            df_opt['combined_signal'] = 0
                            df_opt.loc[(df_opt['macd'].shift(1) < df_opt['signal_line'].shift(1)) & (df_opt['macd'] >= df_opt['signal_line']) & (df_opt['rsi'] > 50), 'combined_signal'] = 1
                            df_opt.loc[(df_opt['macd'].shift(1) > df_opt['signal_line'].shift(1)) & (df_opt['macd'] <= df_opt['signal_line']) & (df_opt['rsi'] < 50), 'combined_signal'] = -1

                            # Drop rows with NaN values created by indicators
                            df_opt.dropna(inplace=True)

                            # Initialize backtesting variables
                            initial_balance = 10000
                            balance = initial_balance
                            position = 'none' # 'none', 'long', 'short'
                            entry_price = 0
                            entry_date = None
                            trades = []

                            # Iterate through the DataFrame
                            if df_opt is None or df_opt.empty:
                                print("DataFrame is empty after dropping NaNs. Cannot perform backtesting.")
                                total_profit_loss = initial_balance - balance # Should be 0
                                total_trades = 0
                                win_rate = 0
                                sharpe_ratio = 0
                                max_drawdown = 0

                            else:
                                for i in range(len(df_opt)):
                                    current_date = df_opt.index[i]
                                    current_price = df_opt['Close'].iloc[i]
                                    current_signal = df_opt['combined_signal'].iloc[i]

                                    # Check for stop loss or take profit conditions first if a position is open
                                    if position == 'long':
                                        stop_loss_price = entry_price * (1 - sl_pct)
                                        take_profit_price = entry_price * (1 + tp_pct)

                                        if current_price <= stop_loss_price:
                                            pnl = (stop_loss_price - entry_price)
                                            balance += pnl
                                            trades.append({'pnl': pnl, 'exit_date': current_date})
                                            position = 'none'
                                            entry_price = 0
                                            entry_date = None

                                        elif current_price >= take_profit_price:
                                            pnl = (take_profit_price - entry_price)
                                            balance += pnl
                                            trades.append({'pnl': pnl, 'exit_date': current_date})
                                            position = 'none'
                                            entry_price = 0
                                            entry_date = None

                                    elif position == 'short':
                                        stop_loss_price = entry_price * (1 + sl_pct)
                                        take_profit_price = entry_price * (1 - tp_pct)

                                        if current_price >= stop_loss_price:
                                            pnl = (entry_price - stop_loss_price)
                                            balance += pnl
                                            trades.append({'pnl': pnl, 'exit_date': current_date})
                                            position = 'none'
                                            entry_price = 0
                                            entry_date = None

                                        elif current_price <= take_profit_price:
                                            pnl = (entry_price - take_profit_price)
                                            balance += pnl
                                            trades.append({'pnl': pnl, 'exit_date': current_date})
                                            position = 'none'
                                            entry_price = 0
                                            entry_date = None

                                    # Enter trade based on signal if no position is open and no exit occurred in this bar
                                    if current_signal == 1 and position == 'none':
                                        position = 'long'
                                        entry_price = current_price
                                        entry_date = current_date

                                    elif current_signal == -1 and position == 'none':
                                        position = 'short'
                                        entry_price = current_price
                                        entry_date = current_date

                                # Calculate performance metrics
                                total_trades = len(trades)
                                total_profit_loss = balance - initial_balance
                                winning_trades = [trade for trade in trades if trade['pnl'] > 0]
                                num_winning_trades = len(winning_trades)
                                win_rate = (num_winning_trades / total_trades) * 100 if total_trades > 0 else 0

                                # Calculate Sharpe Ratio
                                if len(trades) > 1:
                                    equity_curve = pd.Series(initial_balance + np.cumsum([trade['pnl'] for trade in trades]), index=[trade['exit_date'] for trade in trades])
                                    first_signal_date = df_opt['combined_signal'].first_valid_index()
                                    if first_signal_date is not None:
                                        equity_curve = pd.concat([pd.Series(initial_balance, index=[first_signal_date]), equity_curve]).sort_index()
                                        daily_returns = equity_curve.pct_change().dropna()
                                        if daily_returns.std() != 0:
                                            sharpe_ratio = (daily_returns.mean() / daily_returns.std()) * np.sqrt(252)
                                        else:
                                            sharpe_ratio = 0.0
                                    else:
                                        sharpe_ratio = 0.0
                                else:
                                     sharpe_ratio = 0.0

                                # Calculate Maximum Drawdown
                                balance_history = [initial_balance]
                                current_balance = initial_balance
                                for trade in trades:
                                    current_balance += trade['pnl']
                                    balance_history.append(current_balance)

                                peak_balance = -np.inf
                                max_drawdown = 0

                                for balance_val in balance_history:
                                    if balance_val > peak_balance:
                                        peak_balance = balance_val
                                    drawdown_val = (peak_balance - balance_val) / peak_balance if peak_balance != 0 else 0
                                    if drawdown_val > max_drawdown:
                                        max_drawdown = drawdown_val


                            # Store results
                            results_combined_opt.append({
                                'fast_period': fast_period,
                                'slow_period': slow_period,
                                'signal_period': signal_period,
                                'rsi_timeperiod': rsi_timeperiod,
                                'stop_loss_pct': sl_pct,
                                'take_profit_pct': tp_pct,
                                'initial_balance': initial_balance,
                                'final_balance': balance,
                                'total_profit_loss': total_profit_loss,
                                'total_trades': total_trades,
                                'win_rate': win_rate,
                                'sharpe_ratio': sharpe_ratio,
                                'max_drawdown': max_drawdown
                            })
                            print(f"Finished backtesting. Total P/L: {total_profit_loss:.2f}")
                            print("-" * 50)

# Print a summary of the results
print("\n--- Summary of Combined Strategy Optimization Runs ---")
print(f"Completed {len(results_combined_opt)} backtesting runs.")
```
