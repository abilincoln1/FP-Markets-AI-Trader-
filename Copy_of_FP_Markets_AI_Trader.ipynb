```python
# ========== SETUP & DATA LOADING ========== #
!pip install pandas numpy tensorflow ta-lib matplotlib
import pandas as pd
import numpy as np
import talib
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Load your historical data (upload EURUSD_H1.csv first)
# Assuming EURUSD.csv is available from the initial upload and using it as H1 data for demonstration
try:
    df = pd.read_csv("/content/EURUSD.csv", parse_dates=['Date'])
    # Rename 'Date' to 'time' and price columns to match the script's expectations
    df = df.rename(columns={'Date': 'time', 'Open': 'open', 'High': 'high', 'Low': 'low', 'Close': 'close'})
    print(" Data loaded successfully from /content/EURUSD.csv! Samples:")
    print(df[['time', 'open', 'high', 'low', 'close']].head())
except FileNotFoundError:
    print("Error: /content/EURUSD.csv not found. Please ensure the file is uploaded.")
    # Fallback or error handling if the file is not there
    df = None # Set df to None to indicate data loading failed
```

```python
# ========== FEATURE ENGINEERING ========== #
# Calculate indicators
if df is not None: # Check if data was loaded successfully
    df['rsi'] = talib.RSI(df['close'], 14)
    df['macd'], df['signal'], _ = talib.MACD(df['close'], 12, 26, 9)
    df['upper_bb'], _, df['lower_bb'] = talib.BBANDS(df['close'], 20)

    # Create target (1 if next candle is up, 0 if down)
    df['target'] = np.where(df['close'].shift(-1) > df['close'], 1, 0)

    # Normalize data
    scaler = MinMaxScaler()
    features = ['close', 'rsi', 'macd', 'upper_bb', 'lower_bb']
    # Ensure columns exist before scaling and dropping NaNs
    existing_features = [f for f in features if f in df.columns]
    df[existing_features] = scaler.fit_transform(df[existing_features])

    df = df.dropna()
    print(" Features prepared!")
else:
    print("Feature engineering skipped due to data loading error.")
```

```python
# ========== MODEL TRAINING ========== #
if df is not None and not df.empty: # Check if data was loaded and is not empty after dropping NaNs
    # Create sequences for LSTM
    def create_sequences(data, seq_length=50):
        X, y = [], []
        # Ensure we have enough data for sequences
        if len(data) < seq_length:
            print(f"Not enough data to create sequences with length {seq_length}.")
            return np.array(X), np.array(y)

        # Ensure features list is accessible
        features = ['close', 'rsi', 'macd', 'upper_bb', 'lower_bb']
        existing_features = [f for f in features if f in data.columns]
        if len(existing_features) != len(features):
            print(f"Missing required features for sequence creation inside function. Available features: {data.columns.tolist()}")
            return np.array(X), np.array(y)

        # Iterate through the data to create sequences
        for i in range(len(data) - seq_length):
            # Get the window of data for the sequence
            sequence_window = data.iloc[i:i+seq_length]

            # Select only the existing features for X from the sequence window
            X.append(sequence_window[existing_features].values)

            # Access the target value at the end of the sequence window for y
            # Use .iloc[seq_length-1] to get the target corresponding to the last bar in the window
            y.append(data['target'].iloc[i+seq_length-1])


        return np.array(X), np.array(y)

    features = ['close', 'rsi', 'macd', 'upper_bb', 'lower_bb'] # Redefine features list as it was local to the previous cell
    # Ensure all required features exist in the dataframe before calling the function
    existing_features = [f for f in features if f in df.columns]
    if len(existing_features) != len(features):
        print(f"Missing required features before sequence creation. Available features: {df.columns.tolist()}")
        X, y = np.array([]), np.array([]) # Set X and y to empty arrays to prevent errors
    else:
        # Pass the entire df to create_sequences
        X, y = create_sequences(df, seq_length=50)


    if X.shape[0] > 0: # Check if sequences were created
        # Reshape X using the number of existing features
        X = X.reshape(X.shape[0], X.shape[1], len(existing_features))  # Reshape for LSTM

        # Build and train model
        model = Sequential([
            LSTM(50, return_sequences=True, input_shape=(X.shape[1], X.shape[2])),
            LSTM(50),
            Dense(1, activation='sigmoid')
        ])
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        print(" Training model...")
        history = model.fit(X, y, epochs=10, batch_size=32, validation_split=0.2, verbose=1)
        print(" Model trained!")
    else:
        print("Model training skipped as no sequences were created.")
else:
    print("Model training skipped due to data loading or feature engineering errors.")
    model = None # Set model to None if training is skipped
```

```python
# ========== BAR REPLAY TESTING ========== #
def run_bar_replay(data, model, start_idx=1000, end_idx=2000, buy_threshold=0.55, sell_threshold=0.45):
    balance = 10000  # Starting balance
    position = None
    trades = []

    # Ensure end_idx does not exceed the data length
    end_idx = min(end_idx, len(data))
    # Ensure start_idx is valid
    start_idx = max(start_idx, 50) # Need at least 50 candles for the sequence

    # Set up plot
    plt.figure(figsize=(14, 7))
    plt.title(f"LIVE BAR REPLAY - Forex AI Trading (Buy Threshold: {buy_threshold}, Sell Threshold: {sell_threshold})")
    plt.xlabel("Time")
    plt.ylabel("Price")
    plt.grid(True) # Add grid for better readability

    # Plot the initial segment of data
    plt.plot(data['time'].iloc[start_idx-50:start_idx], data['close'].iloc[start_idx-50:start_idx], 'b-', linewidth=1)

    # --- Trading Thresholds (Using function arguments) ---
    # buy_threshold = 0.55 # Now a function argument
    # sell_threshold = 0.45 # Now a function argument
    # -------------------------------------

    for i in range(start_idx, end_idx):
        current_price = data['close'].iloc[i]
        current_time = data['time'].iloc[i]

        # Get last 50 candles for prediction
        # Ensure we have enough data for the sequence starting from i-50
        if i < 50:
            continue # Skip if not enough data for the initial sequence

        # Ensure 'features' list is available in this scope
        if 'features' not in locals():
             print("Error: 'features' list not found. Skipping backtest.")
             return [] # Return empty trades list if features is missing

        seq_data = data[features].iloc[i-50:i].values
        seq_data = seq_data.reshape(1, 50, len(features))

        # Get model prediction
        pred = model.predict(seq_data, verbose=0)[0][0]

        # Trading logic
        if pred > buy_threshold and not position:  # Buy signal with adjusted threshold
            position = {
                'entry_price': current_price,
                'entry_time': current_time,
                'position': 'long',
                'entry_index': i
            }
            plt.plot(current_time, current_price, 'g^', markersize=10, label='Buy Signal' if not any(l.get_label() == 'Buy Signal' for l in plt.gca().lines) else "")  # Green triangle for buy
            # print(f"BUY at {current_time}: {current_price:.5f} (Pred: {pred:.4f})") # Added prediction to print


        elif position and (pred < sell_threshold or i == end_idx-1):  # Sell signal with adjusted threshold or end of data
            if position['position'] == 'long': # Only exit if we are in a long position
                pnl = (current_price - position['entry_price']) * 10000  # P&L calculation (assuming 1 unit size and standard lot)
                balance += pnl
                trades.append({
                    'entry': position['entry_price'],
                    'exit': current_price,
                    'pnl': pnl,
                    'duration': i - position['entry_index']
                })
                plt.plot(current_time, current_price, 'rv', markersize=10, label='Sell Signal' if not any(l.get_label() == 'Sell Signal' for l in plt.gca().lines) else "")  # Red triangle for sell
                # print(f"SELL at {current_time}: {current_price:.5f} | PnL: {pnl:.2f} (Pred: {pred:.4f})") # Added prediction to print
                position = None

        # Update plot with the current bar
        # Ensure there is a previous bar to plot from (i-1)
        if i > start_idx-50: # Start plotting from the beginning of the data shown
             plt.plot(data['time'].iloc[i-1:i+1], data['close'].iloc[i-1:i+1], 'b-', linewidth=1) # Plot the segment from the previous bar to the current bar


        # Update text and pause
        if i % 10 == 0 or i == end_idx - 1: # Update every 10 candles or at the end
             # Remove previous text
            for txt in plt.gca().texts:
                txt.remove()

            plt.text(0.01, 0.95, f"Balance: ${balance:,.2f}", transform=plt.gca().transAxes)
            plt.text(0.01, 0.90, f"Prediction: {pred:.4f}",
                     transform=plt.gca().transAxes)
            plt.text(0.01, 0.85, f"Current Price: {current_price:.5f}",
                     transform=plt.gca().transAxes)
            if position:
                plt.text(0.01, 0.80, f"Position: LONG (Entry: {position['entry_price']:.5f})", transform=plt.gca().transAxes)
            else:
                 plt.text(0.01, 0.80, f"Position: FLAT", transform=plt.gca().transAxes)


            plt.pause(0.001) # Reduced pause time


    # Ensure the final plot is displayed
    plt.legend()
    plt.show()

    # Print results
    print("\n=== BACKTEST RESULTS ===")
    print(f"Starting Balance: $10,000")
    print(f"Final Balance: ${balance:,.2f}")
    if trades:
        win_rate = sum(1 for t in trades if t['pnl'] > 0) / len(trades)
        total_pnl = sum(t['pnl'] for t in trades)
        avg_profit_per_trade = total_pnl / len(trades) if len(trades) > 0 else 0
        print(f"Trades Executed: {len(trades)}")
        print(f"Win Rate: {win_rate:.1%}")
        print(f"Total PnL: ${total_pnl:,.2f}")
        print(f"Avg Profit per Trade: ${avg_profit_per_trade:,.2f}")

    else:
        print("No trades were executed")

    return trades

if 'df' in locals() and df is not None and not df.empty and 'model' in locals() and model is not None:
    print(" Starting Bar Replay Test...")
    # Example usage with default thresholds
    # trades = run_bar_replay(df, model)

    # Example of running with different thresholds for optimization
    # We will run the backtest with the adjusted thresholds (0.55 and 0.45) as agreed
    trades = run_bar_replay(df, model, buy_threshold=0.55, sell_threshold=0.45)

else:
    print("Skipping bar replay test due to missing data or model.")
```

```python
import pandas as pd
import numpy as np
import talib
from sklearn.preprocessing import MinMaxScaler
from tqdm import tqdm # Import tqdm for progress bar
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.model_selection import TimeSeriesSplit
import matplotlib.pyplot as plt # Import for plotting potentially

# Re-load the data. Assuming EURUSD.csv is the intended data for the model.
try:
    df = pd.read_csv("/content/EURUSD.csv", parse_dates=['Date'])
    # Rename 'Date' to 'time' and price columns to match the script's expectations
    # Corrected High and Low based on previous observation
    df = df.rename(columns={'Date': 'time', 'Open': 'open', 'High': 'high', 'Low': 'low', 'Close': 'close'})
    print(" Data re-loaded successfully from /content/EURUSD.csv! Samples:")
    display(df[['time', 'open', 'high', 'low', 'close']].head())
except FileNotFoundError:
    print("Error: /content/EURUSD.csv not found. Please ensure the file is uploaded.")
    df = None # Set df to None to indicate data loading failed

if df is not None and not df.empty:
    # Re-apply feature engineering including the added indicators from the previous successful step
    print("Re-applying feature engineering...")
    df['rsi'] = talib.RSI(df['close'], 14)
    df['macd'], df['signal'], _ = talib.MACD(df['close'], 12, 26, 9)
    df['upper_bb'], _, df['lower_bb'] = talib.BBANDS(df['close'], 20)

    # Add new technical indicators
    df['sma_20'] = talib.SMA(df['close'], timeperiod=20)
    df['sma_50'] = talib.SMA(df['close'], timeperiod=50)
    df['ema_20'] = talib.EMA(df['close'], timeperiod=20)
    df['ema_50'] = talib.EMA(df['close'], timeperiod=50)
    # Ensure High and Low are correctly passed to STOCH
    df['stoch_k'], df['stoch_d'] = talib.STOCH(df['high'], df['low'], df['close'],
                                            fastk_period=14, slowk_period=3, slowd_period=3)
    df['adx'] = talib.ADX(df['high'], df['low'], df['close'], timeperiod=14)


    # Create target (1 if next candle is up, 0 if down)
    df['target'] = np.where(df['close'].shift(-1) > df['close'], 1, 0)

    # Define the updated features list
    features = ['close', 'rsi', 'macd', 'upper_bb', 'lower_bb',
                'sma_20', 'sma_50', 'ema_20', 'ema_50', 'stoch_k', 'stoch_d', 'adx']

    # Normalize data
    scaler = MinMaxScaler()
    # Ensure columns exist before scaling and dropping NaNs
    existing_features = [f for f in features if f in df.columns]
    if len(existing_features) < len(features):
        missing = list(set(features) - set(existing_features))
        print(f"Warning: Missing some expected features after calculation: {missing}")
        # Update features list to only include existing ones for subsequent steps
        features = existing_features


    if existing_features: # Only scale if there are features to scale
        df[existing_features] = scaler.fit_transform(df[existing_features])

    # Drop NaNs again after adding new indicators
    initial_rows = len(df)
    df = df.dropna()
    rows_after_dropna = len(df)
    print(f"Dropped {initial_rows - rows_after_dropna} rows with NaNs after adding new indicators.")
    print(" Features prepared with new indicators!")
    # print first few rows and columns to verify
    display(df[features].head())

    # Re-create sequences for LSTM
    def create_sequences(data, seq_length=50, features_list=None):
        X, y = [], []
        if features_list is None:
            features_list = data.columns.tolist() # Use all columns if not specified

        # Ensure we have enough data for sequences
        if len(data) < seq_length:
            # print(f"Not enough data to create sequences with length {seq_length}. Data length: {len(data)}")
            return np.array(X), np.array(y)

        # Ensure all required features exist in the data
        if not all(f in data.columns for f in features_list):
             missing = list(set(features_list) - set(data.columns))
             print(f"Missing required features for sequence creation: {missing}")
             return np.array(X), np.array(y)


        # Iterate through the data to create sequences
        for i in range(len(data) - seq_length): # No tqdm here to avoid clutter during walk-forward
            # Get the window of data for the sequence, selecting only specified features
            sequence_window = data[features_list].iloc[i:i+seq_length]
            X.append(sequence_window.values)

            # Access the target value at the end of the sequence window for y
            # Use .iloc[seq_length-1] to get the target corresponding to the last bar in the window
            y.append(data['target'].iloc[i+seq_length-1])


        return np.array(X), np.array(y)

    print("Data and features prepared for walk-forward validation.")

    # Define the LSTM model architecture
    def build_lstm_model(input_shape):
        model = Sequential([
            LSTM(50, return_sequences=True, input_shape=input_shape),
            LSTM(50),
            Dense(1, activation='sigmoid')
        ])
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        return model

    # --- Walk-Forward Optimization Setup ---
    # Define walk-forward parameters
    initial_train_size = int(len(df) * 0.7) # Use first 70% for initial training
    test_window_size = int(len(df) * 0.05)  # Use 5% of data for each test window

    # Ensure there are enough data points for at least one split with sequences
    if len(df) - initial_train_size - 50 < test_window_size:
         print(f"Not enough data for walk-forward split with sequence length 50.")
         print(f"Total data points after NaNs: {len(df)}")
         print(f"Initial train size: {initial_train_size}")
         print(f"Remaining data after initial train: {len(df) - initial_train_size}")
         print(f"Required for one test window + sequence: {test_window_size + 50}")
         print("Skipping walk-forward validation.")
         walk_forward_results = [] # Initialize empty list
    else:
        # Use TimeSeriesSplit for walk-forward validation
        # n_splits is the number of test windows
        tscv = TimeSeriesSplit(n_splits=(len(df) - initial_train_size - 50) // test_window_size,
                                 initial_training_size=initial_train_size,
                                 test_size=test_window_size)

        walk_forward_results = [] # Store evaluation metrics for each fold

        print(f"\nStarting Walk-Forward Validation with {tscv.get_n_splits()} splits...")

        fold = 0
        # Iterate through the time series splits
        for train_index, test_index in tscv.split(df):
            fold += 1
            print(f"\n--- Fold {fold} ---")

            # Split data into training and testing sets for the current fold
            train_data = df.iloc[train_index]
            test_data = df.iloc[test_index]

            # Create sequences for the current split
            print(f"Creating sequences for Fold {fold} Train ({len(train_data)} samples)...")
            # Ensure we pass the correct subset of data for sequence creation
            X_train, y_train = create_sequences(train_data, seq_length=50, features_list=features)

            print(f"Creating sequences for Fold {fold} Test ({len(test_data)} samples)...")
            # For test data, ensure we have enough preceding data for the first sequence
            # This might require including some data from the end of the training set
            # A simpler approach with TimeSeriesSplit is to slice the already created X and y
            # based on the indices returned by tscv.split.
            # However, create_sequences drops initial rows due to sequence length and NaNs.
            # A more robust approach is to apply create_sequences to the *entire* df
            # and then slice the resulting X and y based on tscv indices, after adjusting for the sequence offset.

            # Let's recreate sequences on the entire df first, outside the loop,
            # and then slice X and y inside the loop. This is more typical for TimeSeriesSplit.
            # Need to re-run sequence creation on the full df if it wasn't done properly before.
            # Let's assume df still contains the processed data after dropping NaNs.

            # Re-create sequences on the full processed dataframe (after NaNs are dropped)
            # This part should ideally be outside the loop if done on the full df once.
            # But let's keep it here for now to ensure data is current for the loop.
            # This is NOT the standard way to use TimeSeriesSplit indices with sequence data.
            # Correct way: create sequences on full df, then use adjusted indices.
            # Let's adjust the indices approach.

            # Corrected approach: Create sequences on the full df outside the loop.
            # Then, inside the loop, map tscv indices to the sequence indices.

            print("Creating sequences on the full dataset for slicing...")
            # Ensure df is the dataframe after dropping NaNs and scaling.
            X_full, y_full = create_sequences(df, seq_length=50, features_list=features)
            print(f"Full sequences created. X_full shape: {X_full.shape}, y_full shape: {y_full.shape}")

            if X_full.shape[0] == 0:
                 print("Failed to create sequences on the full dataset. Skipping walk-forward.")
                 walk_forward_results = []
                 break # Exit the walk-forward loop


            # Adjust indices from TimeSeriesSplit to match the indices of the sequence data (X_full, y_full)
            # The sequence data at index `i` corresponds to the original dataframe data ending at index `i + seq_length - 1`.
            # So, the original dataframe index `j` corresponds to the sequence data index `j - seq_length + 1`.
            # We also need to account for rows dropped due to NaNs at the start of the original df.
            # Let's find the starting index of the sequence data relative to the original df.
            original_df_start_index = df.index[0] # First index in the df after dropping NaNs
            sequence_start_index_in_original_df = original_df_start_index + 50 -1 # The index in original df that is the last bar of the first sequence

            # Map train_index and test_index (which refer to indices in the *current* df)
            # to the indices of X_full and y_full.
            # The first sequence in X_full corresponds to df.iloc[0:50], its target y_full[0] is df['target'].iloc[49].
            # The sequence ending at df.iloc[i] corresponds to X_full[i-50]. Its target is y_full[i-50] which is df['target'].iloc[i].
            # So, if the original index in the current df is `idx`, the corresponding sequence is at `idx - 50`.
            # The target for that sequence is at `idx`.

            # Let's redefine how to get train/test sequences based on tscv indices
            # The tscv indices point to the rows in the *current* df after NaNs are dropped.
            # We need sequences *ending* at these indices.
            # A sequence ending at index `i` in `df` is `df.iloc[i-49:i+1]`.
            # The target for this sequence is `df['target'].iloc[i+1]`.
            # This is slightly different from the create_sequences function definition which creates
            # sequences ending at i+seq_length-1 and targets at i+seq_length-1.
            # Let's align with the create_sequences definition for consistency.

            # Adjusted approach: Create sequences on the full df. Then, the tscv indices (relative to df)
            # can be mapped to the indices of the sequence data.
            # If the tscv test_index starts at `test_start_idx_df` in `df`, the corresponding
            # sequence data `X_full` indices start at `test_start_idx_df - 50`.
            # The target data `y_full` indices start at `test_start_idx_df - 50`.

            # Let's simplify and use the original create_sequences logic per split for now,
            # even though it's less efficient. A robust walk-forward for sequences is complex.
            # Reverting to creating sequences per split for clarity in this step.
            # This means `create_sequences` needs to handle the start index correctly for test sets.

            # Let's rethink the loop with the original create_sequences function.
            # train_data and test_data are slices of df.
            # create_sequences(train_data) will create sequences *within* the train_data slice.
            # create_sequences(test_data) will create sequences *within* the test_data slice.
            # This means the test set sequences will use data ONLY from the test_data slice.
            # This is incorrect for time series, as test sequences need lookback from the end of train data.

            # Correct Walk-Forward Approach for Sequences:
            # 1. Split original data (before sequence creation) using TimeSeriesSplit.
            # 2. For each split, take the train data and the test data.
            # 3. To create sequences for the test data, you need the last `seq_length - 1` bars from the training data.
            # 4. Combine the end of train data with the test data slice for sequence creation.

            # Let's implement the correct approach.

            print("Creating sequences for Fold {fold}...")

            # Combine train_data and the necessary lookback from train_data for the test set
            # The test set sequences will start at the beginning of test_data, requiring `seq_length - 1` lookback.
            lookback_data_for_test = train_data.iloc[-(50-1):] if 50 > 1 else train_data.tail(0) # Get the last 49 rows of train_data
            combined_test_data_source = pd.concat([lookback_data_for_test, test_data])


            X_train, y_train = create_sequences(train_data.copy(), seq_length=50, features_list=features)
            # The target for the test set is the actual next bar after the sequence.
            # Let's redefine create_sequences slightly to return sequences and their *next* target.

            def create_sequences_with_next_target(data, seq_length=50, features_list=None):
                 X, y = [], []
                 if features_list is None:
                     features_list = data.columns.tolist()

                 if len(data) < seq_length + 1: # Need seq_length for input + 1 for target
                     # print(f"Not enough data to create sequences with next target. Data length: {len(data)}, Required: {seq_length + 1}")
                     return np.array(X), np.array(y)

                 if not all(f in data.columns for f in features_list):
                     missing = list(set(features_list) - set(data.columns))
                     print(f"Missing required features for sequence creation with next target: {missing}")
                     return np.array(X), np.array(y)

                 for i in range(len(data) - seq_length):
                     sequence_window = data[features_list].iloc[i:i+seq_length]
                     X.append(sequence_window.values)
                     # The target is the 'target' value of the bar *after* the sequence
                     y.append(data['target'].iloc[i+seq_length])

                 return np.array(X), np.array(y)

            # Use the updated create_sequences function
            print(f"Creating sequences for Fold {fold} Train ({len(train_data)} samples)...")
            X_train, y_train = create_sequences_with_next_target(train_data.copy(), seq_length=50, features_list=features)

            print(f"Creating sequences for Fold {fold} Test ({len(combined_test_data_source)} samples including lookback)...")
            # Create sequences on the combined data source, but only take the sequences corresponding to the actual test_data indices
            X_combined_test, y_combined_test = create_sequences_with_next_target(combined_test_data_source.copy(), seq_length=50, features_list=features)

            # X_combined_test and y_combined_test contain sequences starting from the lookback period.
            # We only want the sequences whose target is within the actual test_data period.
            # The sequences are created for indices `i` from 0 up to `len(data) - seq_length - 1`.
            # The target for sequence `i` is at index `i + seq_length`.
            # In `combined_test_data_source`, the actual test data starts at index `len(lookback_data_for_test)`.
            # We want sequences whose target is at index `j` where `j >= len(lookback_data_for_test)`.
            # This corresponds to sequence index `i = j - seq_length`.
            # So, the test sequences in `X_combined_test` start at index `len(lookback_data_for_test) - seq_length`.
            # And the corresponding targets in `y_combined_test` start at the same index.

            # Let's recalculate the start index in the combined data source for the test set targets.
            test_data_start_index_in_combined = len(lookback_data_for_test)
            # The target for a sequence starting at index `i` in `combined_test_data_source` is at `i + seq_length`.
            # We want targets that are part of the original `test_data`.
            # The original `test_data` corresponds to indices from `test_data_start_index_in_combined` onwards in `combined_test_data_source`.
            # The first target in `y_combined_test` that corresponds to the actual test set in `test_data`
            # is the one whose corresponding bar in `combined_test_data_source` is at index `test_data_start_index_in_combined`.
            # The sequence ending at index `test_data_start_index_in_combined - 1` has its target at `test_data_start_index_in_combined`.
            # The index of this target in `y_combined_test` is `(test_data_start_index_in_combined) - seq_length`. This seems incorrect.

            # Let's simplify the slicing based on the fact that TimeSeriesSplit gives indices relative to the original data.
            # The original `df` has indices from 0 to len(df)-1 after dropping initial NaNs.
            # TimeSeriesSplit `train_index` and `test_index` refer to these indices.
            # The sequences `X_full` and `y_full` are created on this `df`.
            # The sequence `X_full[i]` corresponds to data from `df.iloc[i : i + 50]`.
            # The target `y_full[i]` corresponds to `df['target'].iloc[i + 50]`.
            # So, if the tscv `train_index` is `[0, 1, ..., k]`, the corresponding sequences are `X_full[0:k-49]` and targets `y_full[0:k-49]`.
            # If the tscv `test_index` is `[m, m+1, ..., n]`, the corresponding sequences are `X_full[m-49:n-49]` and targets `y_full[m-49:n-49]`.
            # This requires careful index mapping and ensuring `m-49 >= 0`.

            # Let's go back to creating sequences per split but ensure lookback is included for the test set.

            print(f"Creating sequences for Fold {fold} Train ({len(train_data)} samples)...")
            X_train, y_train = create_sequences_with_next_target(train_data.copy(), seq_length=50, features_list=features)

            print(f"Creating sequences for Fold {fold} Test ({len(test_data)} samples)...")
            # To create test sequences that start at the beginning of test_data and include lookback:
            # We need `seq_length` worth of data *before* the start of `test_data`.
            # This data comes from the end of `train_data`.
            lookback_for_test = train_data.tail(50 - 1) if 50 > 1 else train_data.tail(0)
            test_data_with_lookback = pd.concat([lookback_for_test, test_data])

            # Now create sequences on `test_data_with_lookback`.
            # The sequences created will start from the beginning of this combined data.
            # We only care about the sequences whose *first element* is within the actual `test_data` part.
            # A sequence starting at index `i` in `test_data_with_lookback` uses data from `i` to `i + seq_length - 1`.
            # Its target is at `i + seq_length`.
            # The actual `test_data` starts at index `len(lookback_for_test)` in `test_data_with_lookback`.
            # We need sequences starting from index `len(lookback_for_test)`.
            # The sequences `X_combined_test` and targets `y_combined_test` will be created for indices `i` from 0 to `len(test_data_with_lookback) - seq_length - 1`.
            # We need to slice these results to get only the test set sequences.
            # The sequences corresponding to the actual test data start where the input window no longer fully overlaps with the lookback data.
            # The first sequence that uses data *from* the test_data starts when its first element is at index `len(lookback_for_test)`.
            # The index of this sequence in `X_combined_test` is `len(lookback_for_test)`. This logic feels simpler.

            X_test_full, y_test_full = create_sequences_with_next_target(test_data_with_lookback.copy(), seq_length=50, features_list=features)

            # Slice X_test_full and y_test_full to get only the sequences corresponding to the actual test set.
            # The actual test data in `test_data_with_lookback` starts at index `len(lookback_for_test)`.
            # A sequence starting at index `i` in `test_data_with_lookback` has its target at `i + seq_length`.
            # We want sequences whose target is at index `j` in `test_data_with_lookback` where `j >= len(lookback_for_test)`.
            # This means sequence index `i = j - seq_length`.
            # The first sequence index we want is when `j = len(lookback_for_test)`. So, `i = len(lookback_for_test) - seq_length`. This is still confusing.

            # Let's simplify. `X_test_full` contains sequences created from `test_data_with_lookback`.
            # The number of sequences created is `len(test_data_with_lookback) - seq_length`.
            # The sequences that correspond to the actual test set are the last `len(test_data)` sequences.
            # This is because each sequence corresponds to predicting the *next* target. The sequence at index `i` in `X_test_full`
            # predicts the target at index `i + seq_length` in `test_data_with_lookback`.
            # We want the sequences that predict targets within the actual `test_data` indices.
            # The actual `test_data` indices in `test_data_with_lookback` are from `len(lookback_for_test)` to `len(test_data_with_lookback) - 1`.
            # So we want sequences `i` such that `i + seq_length >= len(lookback_for_test)`.
            # This means `i >= len(lookback_for_test) - seq_length`.
            # The sequences in `X_test_full` corresponding to the actual test data start at index `len(lookback_for_test) - seq_length`. This index can be negative.

            # Let's use the indices relative to the start of the combined data source.
            # The number of lookback bars is `seq_length - 1`.
            # The test data starts after the lookback bars.
            # The sequences in `X_test_full` that are purely based on the lookback data are the first `len(lookback_for_test) - (seq_length -1)` sequences.
            # The sequences in `X_test_full` whose *target* is within the actual test data range starts after the lookback data.
            # The first target in `test_data_with_lookback` that is part of the actual test set is at index `len(lookback_for_test)`.
            # The sequence corresponding to this target is at index `len(lookback_for_test) - seq_length`.

            # Let's try a simpler slicing: X_test will be the last `len(test_data)` sequences from X_test_full.
            # This assumes each sequence predicts the target *immediately* following the sequence.
            # This seems correct based on the `create_sequences_with_next_target` definition.

            # Check if enough sequences were created from test_data_with_lookback
            if len(X_test_full) >= len(test_data):
                X_test = X_test_full[-len(test_data):]
                y_test = y_test_full[-len(test_data):]
                print(f"Fold {fold} Train sequences: {X_train.shape}, Test sequences: {X_test.shape}")
            else:
                print(f"Not enough sequences created for Fold {fold} test set. Needed {len(test_data)}, got {len(X_test_full)}. Skipping fold.")
                continue # Skip this fold if not enough test sequences


            if X_train.shape[0] > 0 and X_test.shape[0] > 0:
                # Build and train the model for the current fold
                model = build_lstm_model(input_shape=(X_train.shape[1], X_train.shape[2]))

                print(f"Training model for Fold {fold}...")
                history = model.fit(
                    X_train, y_train,
                    epochs=10, # Fixed epochs for walk-forward
                    batch_size=32, # Fixed batch size
                    verbose=0 # Keep training output clean
                )

                # Evaluate the trained model on the test set
                print(f"Evaluating model for Fold {fold}...")
                loss, accuracy = model.evaluate(X_test, y_test, verbose=0)

                # Store the evaluation metrics
                walk_forward_results.append({'fold': fold, 'loss': loss, 'accuracy': accuracy})
                print(f"Fold {fold} - Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}")

            else:
                print(f"Skipping training and evaluation for Fold {fold} due to insufficient training or test sequences.")


        # --- Summarize Walk-Forward Results ---
        if walk_forward_results:
            results_df = pd.DataFrame(walk_forward_results)
            print("\nWalk-Forward Validation Summary:")
            display(results_df)

            print("\nAverage Performance across Folds:")
            print(results_df[['loss', 'accuracy']].mean())

            print("\nStandard Deviation of Performance across Folds:")
            print(results_df[['loss', 'accuracy']].std())

            # Optional: Plot results
            plt.figure(figsize=(10, 6))
            plt.plot(results_df['fold'], results_df['accuracy'], marker='o')
            plt.title('Walk-Forward Test Accuracy per Fold')
            plt.xlabel('Fold Number')
            plt.ylabel('Accuracy')
            plt.grid(True)
            plt.show()

        else:
            print("\nNo walk-forward results to summarize.")


else:
    print("Data loading or feature engineering failed, skipping walk-forward validation.")
```

## Data Preparation for Model Training

### Subtask:
Load daily EURUSD data, perform feature engineering, and create sequences for model training.

**Reasoning**:
Re-load the daily EURUSD data, apply feature engineering including new indicators, normalize data, handle NaNs, and create sequences to make data ready for model training and exploration.

```python
import pandas as pd
import numpy as np
import talib
from sklearn.preprocessing import MinMaxScaler
from tqdm import tqdm # Import tqdm for progress bar

# Re-load the daily EURUSD data
try:
    df_daily = pd.read_csv("/content/EURUSD.csv", parse_dates=['Date'])
    # Rename columns to match the expected format of the original script
    df_daily = df_daily.rename(columns={'Date': 'time', 'Open': 'open', 'High': 'high', 'Low': 'low', 'Close': 'close'})
    print(" Daily EURUSD data re-loaded successfully!")
    display(df_daily.head())
except FileNotFoundError:
    print("Error: /content/EURUSD.csv not found. Please ensure the file is uploaded.")
    df_daily = None # Set to None if loading fails

if df_daily is not None and not df_daily.empty:
    # --- Feature Engineering (reusing logic from previous script) ---
    print("Applying feature engineering for daily data...")
    df_daily['rsi'] = talib.RSI(df_daily['close'], 14)
    df_daily['macd'], df_daily['signal'], _ = talib.MACD(df_daily['close'], 12, 26, 9)
    df_daily['upper_bb'], _, df_daily['lower_bb'] = talib.BBANDS(df_daily['close'], 20)

    # Add new technical indicators (from previous attempts to enhance features)
    df_daily['sma_20'] = talib.SMA(df_daily['close'], timeperiod=20)
    df_daily['sma_50'] = talib.SMA(df_daily['close'], timeperiod=50)
    df_daily['ema_20'] = talib.EMA(df_daily['close'], timeperiod=20)
    df_daily['ema_50'] = talib.EMA(df_daily['close'], timeperiod=50)
    df_daily['stoch_k'], df_daily['stoch_d'] = talib.STOCH(df_daily['high'], df_daily['low'], df_daily['close'],
                                                    fastk_period=14, slowk_period=3, slowd_period=3)
    df_daily['adx'] = talib.ADX(df_daily['high'], df_daily['low'], df_daily['close'], timeperiod=14)


    # Create target (1 if next candle is up, 0 if down) - using shift(1) for next day's close
    df_daily['target'] = np.where(df_daily['close'].shift(-1) > df_daily['close'], 1, 0)

    # Define the features list (updated with new indicators)
    features_daily = ['close', 'rsi', 'macd', 'upper_bb', 'lower_bb',
                      'sma_20', 'sma_50', 'ema_20', 'ema_50', 'stoch_k', 'stoch_d', 'adx']

    # Normalize data
    scaler_daily = MinMaxScaler()
    # Ensure columns exist before scaling and dropping NaNs
    existing_features_daily = [f for f in features_daily if f in df_daily.columns]
    if len(existing_features_daily) < len(features_daily):
        missing = list(set(features_daily) - set(existing_features_daily))
        print(f"Warning: Missing some expected features after calculation: {missing}")
        # Update features list to only include existing ones
        features_daily = existing_features_daily

    if features_daily: # Only scale if there are features to scale
        df_daily[features_daily] = scaler_daily.fit_transform(df_daily[features_daily])

    # Drop NaNs after adding all indicators and target
    initial_rows_daily = len(df_daily)
    df_daily = df_daily.dropna().reset_index(drop=True) # Reset index after dropping NaNs
    rows_after_dropna_daily = len(df_daily)
    print(f"Dropped {initial_rows_daily - rows_after_dropna_daily} rows with NaNs after adding indicators and target.")
    print(" Features prepared for daily data!")
    display(df_daily[features_daily].head())

    # --- Sequence Creation (reusing logic from previous script) ---
    def create_sequences(data, seq_length=50, features_list=None):
        X, y = [], []
        if features_list is None:
            features_list = data.columns.tolist()

        if len(data) < seq_length + 1: # Need seq_length for input + 1 for target
            print(f"Not enough data to create sequences with next target. Data length: {len(data)}, Required: {seq_length + 1}")
            return np.array(X), np.array(y)

        if not all(f in data.columns for f in features_list):
             missing = list(set(features_list) - set(data.columns))
             print(f"Missing required features for sequence creation with next target: {missing}")
             return np.array(X), np.array(y)

        for i in tqdm(range(len(data) - seq_length), desc="Creating sequences"):
            sequence_window = data[features_list].iloc[i:i+seq_length]
            X.append(sequence_window.values)
            # The target is the 'target' value of the bar *after* the sequence
            y.append(data['target'].iloc[i+seq_length])

        return np.array(X), np.array(y)

    print("Creating training sequences for daily data...")
    # Pass the features list to the function
    X_daily, y_daily = create_sequences(df_daily, seq_length=50, features_list=features_daily)

    if X_daily.shape[0] > 0:
        print(f" Daily sequences created. X_daily shape: {X_daily.shape}, y_daily shape: {y_daily.shape}")

    else:
        print(" No daily sequences were created.")

else:
    print("Daily data loading or feature engineering failed.")
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import LSTM, Dense, GRU, Concatenate, Input
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split # Import train_test_split
import itertools # To iterate through hyperparameter combinations

# Ensure X_daily, y_daily, features_daily are available from previous steps and not None
if ('X_daily' in locals() and X_daily.shape[0] > 0 and
    'y_daily' in locals() and y_daily.shape[0] > 0 and
    'features_daily' in locals() and len(features_daily) > 0):

    # Define a function to build the model with hyperparameters as arguments
    # We'll use the Simple_LSTM_V2 architecture as an example for tuning
    def build_tuned_lstm_model(lstm_units=64, dense_units=16, learning_rate=0.001, input_shape=None):
        model = Sequential([
            LSTM(lstm_units, return_sequences=True, input_shape=input_shape),
            LSTM(lstm_units // 2), # Reduce units in the second LSTM
            Dense(dense_units, activation='relu'),
            Dense(1, activation='sigmoid')
        ])
        optimizer = Adam(learning_rate=learning_rate)
        model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
        return model

    # Define a hyperparameter grid to search over
    param_grid = {
        'lstm_units': [32, 64],
        'dense_units': [8, 16],
        'learning_rate': [0.01, 0.001, 0.0001],
        'batch_size': [32, 64],
        'epochs': [15, 20] # Limit epochs for faster exploration
    }

    # Prepare data for tuning (using a fixed train/validation split for each combination)
    X_train_tune, X_val_tune, y_train_tune, y_val_tune = train_test_split(
        X_daily, y_daily, test_size=0.2, shuffle=False # Use shuffle=False for time series data
    )

    best_accuracy = 0
    best_params = {}
    tuning_results = []

    print("Starting Hyperparameter Tuning...")

    # Iterate through all combinations of hyperparameters
    for params in itertools.product(*param_grid.values()):
        current_params = dict(zip(param_grid.keys(), params))
        print(f"\nTraining with params: {current_params}")

        # Build and compile the model with current hyperparameters
        # Ensure input_shape is passed correctly
        model = build_tuned_lstm_model(
            lstm_units=current_params['lstm_units'],
            dense_units=current_params['dense_units'],
            learning_rate=current_params['learning_rate'],
            input_shape=(X_train_tune.shape[1], X_train_tune.shape[2])
        )

        # Train the model
        history = model.fit(
            X_train_tune, y_train_tune,
            epochs=current_params['epochs'],
            batch_size=current_params['batch_size'],
            validation_data=(X_val_tune, y_val_tune),
            verbose=0 # Keep training output clean
        )

        # Evaluate on the validation set
        loss, accuracy = model.evaluate(X_val_tune, y_val_tune, verbose=0)

        print(f"Validation Accuracy: {accuracy:.4f}")

        # Store results
        tuning_results.append({**current_params, 'val_accuracy': accuracy, 'val_loss': loss})

        # Check if this is the best model so far
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_params = current_params
            # Optionally save the best model here

    print("\nHyperparameter Tuning Complete.")
    print(f"Best Validation Accuracy: {best_accuracy:.4f}")
    print(f"Best Hyperparameters: {best_params}")

    # Optionally display all tuning results in a DataFrame
    tuning_results_df = pd.DataFrame(tuning_results)
    print("\nTuning Results Summary:")
    display(tuning_results_df.sort_values(by='val_accuracy', ascending=False))


else:
    print("Skipping hyperparameter tuning as data (X_daily, y_daily) or features (features_daily) are not available or empty.")
    print("Please ensure the data loading, feature engineering, and sequence creation steps were successful for daily data.")
```

## Robust backtesting and validation

### Subtask:
Implement more rigorous backtesting methodologies, including out-of-sample testing and walk-forward optimization, to ensure the strategy's performance is not just a result of overfitting.

**Reasoning**:
Reload the EURUSD data and perform feature engineering and sequence creation based on the previous successful attempts and then implement time series split for walk-forward validation, iterate through the splits, train and evaluate the model, and summarize the results.

```python
import pandas as pd
import numpy as np
import talib
from sklearn.preprocessing import MinMaxScaler
from tqdm import tqdm # Import tqdm for progress bar
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.model_selection import TimeSeriesSplit
import matplotlib.pyplot as plt # Import for plotting potentially

# Re-load the data. Assuming EURUSD.csv is the intended data for the model.
try:
    df = pd.read_csv("/content/EURUSD.csv", parse_dates=['Date'])
    # Rename 'Date' to 'time' and price columns to match the script's expectations
    # Corrected High and Low based on previous observation
    df = df.rename(columns={'Date': 'time', 'Open': 'open', 'High': 'high', 'Low': 'low', 'Close': 'close'})
    print(" Data re-loaded successfully from /content/EURUSD.csv! Samples:")
    display(df[['time', 'open', 'high', 'low', 'close']].head())
except FileNotFoundError:
    print("Error: /content/EURUSD.csv not found. Please ensure the file is uploaded.")
    df = None # Set df to None to indicate data loading failed

if df is not None and not df.empty:
    # Re-apply feature engineering including the added indicators from the previous successful step
    print("Re-applying feature engineering...")
    df['rsi'] = talib.RSI(df['close'], 14)
    df['macd'], df['signal'], _ = talib.MACD(df['close'], 12, 26, 9)
    df['upper_bb'], _, df['lower_bb'] = talib.BBANDS(df['close'], 20)

    # Add new technical indicators
    df['sma_20'] = talib.SMA(df['close'], timeperiod=20)
    df['sma_50'] = talib.SMA(df['close'], timeperiod=50)
    df['ema_20'] = talib.EMA(df['close'], timeperiod=20)
    df['ema_50'] = talib.EMA(df['close'], timeperiod=50)
    # Ensure High and Low are correctly passed to STOCH
    df['stoch_k'], df['stoch_d'] = talib.STOCH(df['high'], df['low'], df['close'],
                                            fastk_period=14, slowk_period=3, slowd_period=3)
    df['adx'] = talib.ADX(df['high'], df['low'], df['close'], timeperiod=14)


    # Create target (1 if next candle is up, 0 if down)
    df['target'] = np.where(df['close'].shift(-1) > df['close'], 1, 0)

    # Define the updated features list
    features = ['close', 'rsi', 'macd', 'upper_bb', 'lower_bb',
                'sma_20', 'sma_50', 'ema_20', 'ema_50', 'stoch_k', 'stoch_d', 'adx']

    # Normalize data
    scaler = MinMaxScaler()
    # Ensure columns exist before scaling and dropping NaNs
    existing_features = [f for f in features if f in df.columns]
    if len(existing_features) < len(features):
        missing = list(set(features) - set(existing_features))
        print(f"Warning: Missing some expected features after calculation: {missing}")
        # Update features list to only include existing ones for subsequent steps
        features = existing_features


    if existing_features: # Only scale if there are features to scale
        df[existing_features] = scaler.fit_transform(df[existing_features])

    # Drop NaNs again after adding new indicators
    initial_rows = len(df)
    df = df.dropna().reset_index(drop=True)
    rows_after_dropna = len(df)
    print(f"Dropped {initial_rows - rows_after_dropna} rows with NaNs after adding new indicators.")
    print(" Features prepared with new indicators!")
    # print first few rows and columns to verify
    display(df[features].head())

    # Re-create sequences for LSTM
    def create_sequences(data, seq_length=50, features_list=None):
        X, y = [], []
        if features_list is None:
            features_list = data.columns.tolist() # Use all columns if not specified

        if len(data) < seq_length + 1: # Need seq_length for input + 1 for target
            # print(f"Not enough data to create sequences with next target. Data length: {len(data)}, Required: {seq_length + 1}")
            return np.array(X), np.array(y)

        if not all(f in data.columns for f in features_list):
             missing = list(set(features_list) - set(data.columns))
             print(f"Missing required features for sequence creation with next target: {missing}")
             return np.array(X), np.array(y)


        # Iterate through the data to create sequences
        for i in range(len(data) - seq_length): # No tqdm here to avoid clutter during walk-forward
            # Get the window of data for the sequence, selecting only specified features
            sequence_window = data[features_list].iloc[i:i+seq_length]
            X.append(sequence_window.values)

            # The target is the 'target' value of the bar *after* the sequence
            y.append(data['target'].iloc[i+seq_length])


        return np.array(X), np.array(y)

    print("Data and features prepared for walk-forward validation.")

    # Define the LSTM model architecture
    def build_lstm_model(input_shape):
        model = Sequential([
            LSTM(50, return_sequences=True, input_shape=input_shape),
            LSTM(50),
            Dense(1, activation='sigmoid')
        ])
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        return model

    # --- Walk-Forward Optimization Setup ---
    # Define walk-forward parameters
    # TimeSeriesSplit in newer scikit-learn versions uses 'max_train_size' instead of 'initial_training_size'
    # And the number of splits is determined by n_splits, which defines the number of test folds.
    # The size of each test fold is determined by test_size.
    # The training set grows with each split.

    n_splits = 5 # Number of walk-forward splits (adjust as needed)
    test_size = int(len(df) * 0.1) # Size of each test set (e.g., 10% of data)
    # The training size for the first split will be len(df) - n_splits * test_size - initial sequence length
    # TimeSeriesSplit automatically handles the growing window.

    # Ensure there are enough data points for the splits
    if len(df) < n_splits * test_size + 50: # Need at least enough data for splits + initial sequence
        print(f"Not enough data for {n_splits} walk-forward splits with test size {test_size} and sequence length 50.")
        print(f"Total data points after NaNs: {len(df)}")
        print(f"Required data points: {n_splits * test_size + 50}")
        print("Skipping walk-forward validation.")
        walk_forward_results = [] # Initialize empty list
    else:
        # Removed initial_training_size as it's not a valid parameter in some TimeSeriesSplit versions
        # The training set size is implicitly determined by n_splits and test_size.
        tscv = TimeSeriesSplit(n_splits=n_splits, test_size=test_size)

        walk_forward_results = [] # Store evaluation metrics for each fold

        print(f"\nStarting Walk-Forward Validation with {tscv.get_n_splits()} splits...")

        fold = 0
        # Iterate through the time series splits
        for train_index, test_index in tscv.split(df):
            fold += 1
            print(f"\n--- Fold {fold} ---")

            # Get the data for the current fold
            train_data_fold = df.iloc[train_index]
            test_data_fold = df.iloc[test_index]

            # Create sequences for the current fold's training data
            # Need to ensure enough data for sequences within the train_data_fold
            if len(train_data_fold) < 50 + 1: # Need sequence length + 1 for target
                print(f"Not enough data in training fold {fold} to create sequences ({len(train_data_fold)} samples). Skipping fold.")
                continue # Skip this fold

            X_train, y_train = create_sequences(train_data_fold.copy(), seq_length=50, features_list=features)

            # Create sequences for the current fold's test data, including necessary lookback from training data
            seq_length = 50 # Redefine seq_length locally for clarity
            lookback_for_test = train_data_fold.tail(seq_length - 1) if seq_length > 1 else train_data_fold.tail(0)
            combined_test_data_source = pd.concat([lookback_for_test, test_data_fold])

            # Create sequences on the combined data source for testing
            # We need enough data in combined_test_data_source to create at least one sequence whose target is in the actual test_data_fold.
            # The actual test_data_fold starts after the lookback period in combined_test_data_source.
            # The first target in the actual test_data_fold is at index `len(lookback_for_test)` within `combined_test_data_source`.
            # The sequence corresponding to this target starts at index `len(lookback_for_test) - seq_length` within `combined_test_data_source`.
            # So, combined_test_data_source needs to have at least `len(lookback_for_test)` + 1 data points for the first test target,
            # which means `len(lookback_for_test) + 1 >= seq_length + 1` for create_sequences to generate the first sequence.

            # Let's check if there's enough data in combined_test_data_source to create sequences that actually cover the test data.
            # The number of sequences created from combined_test_data_source is `len(combined_test_data_source) - seq_length`.
            # We need this number to be at least `len(test_data_fold)`.
            if len(combined_test_data_source) - seq_length < len(test_data_fold):
                 print(f"Not enough data in test fold {fold} (including lookback) to create {len(test_data_fold)} sequences. Got {len(combined_test_data_source) - seq_length}. Skipping fold.")
                 continue # Skip this fold

            X_test_full, y_test_full = create_sequences(combined_test_data_source.copy(), seq_length=seq_length, features_list=features)

            # Slice X_test_full and y_test_full to get only the sequences corresponding to the actual test set.
            # The sequences in X_test_full that correspond to the actual test data are the last `len(test_data_fold)` sequences.
            if len(X_test_full) >= len(test_data_fold):
                X_test = X_test_full[-len(test_data_fold):]
                y_test = y_test_full[-len(test_data_fold):]
                print(f"Fold {fold} Train sequences: {X_train.shape}, Test sequences: {X_test.shape}")
            else:
                # This case should be covered by the previous check, but keeping for safety.
                print(f"Not enough sequences created for Fold {fold} test set after creating from combined data. Needed {len(test_data_fold)}, got {len(X_test_full)}. Skipping fold.")
                continue # Skip this fold


            if X_train.shape[0] > 0 and X_test.shape[0] > 0:
                # Build and train the model for the current fold
                model = build_lstm_model(input_shape=(X_train.shape[1], X_train.shape[2]))

                print(f"Training model for Fold {fold}...")
                history = model.fit(
                    X_train, y_train,
                    epochs=10, # Fixed epochs for walk-forward
                    batch_size=32, # Fixed batch size
                    verbose=0 # Keep training output clean
                )

                # Evaluate the trained model on the test set
                print(f"Evaluating model for Fold {fold}...")
                loss, accuracy = model.evaluate(X_test, y_test, verbose=0)

                # Store the evaluation metrics
                walk_forward_results.append({'fold': fold, 'loss': loss, 'accuracy': accuracy})
                print(f"Fold {fold} - Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}")

            else:
                print(f"Skipping training and evaluation for Fold {fold} due to insufficient training or test sequences.")


        # --- Summarize Walk-Forward Results ---
        if walk_forward_results:
            results_df = pd.DataFrame(walk_forward_results)
            print("\nWalk-Forward Validation Summary:")
            display(results_df)

            print("\nAverage Performance across Folds:")
            print(results_df[['loss', 'accuracy']].mean())

            print("\nStandard Deviation of Performance across Folds:")
            print(results_df[['loss', 'accuracy']].std())

            # Optional: Plot results
            plt.figure(figsize=(10, 6))
            plt.plot(results_df['fold'], results_df['accuracy'], marker='o')
            plt.title('Walk-Forward Test Accuracy per Fold')
            plt.xlabel('Fold Number')
            plt.ylabel('Accuracy')
            plt.grid(True)
            plt.show()

        else:
            print("\nNo walk-forward results to summarize.")


else:
    print("Data loading or feature engineering failed, skipping walk-forward validation.")
```

## Hyperparameter tuning

### Subtask:
Optimize the hyperparameters of the chosen model and the training process (e.g., learning rate, number of epochs, batch size, LSTM units) to find the best configuration.

**Reasoning**:
Define and train alternative model architectures.

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import LSTM, Dense, GRU, Concatenate, Input
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split # Import train_test_split
import itertools # To iterate through hyperparameter combinations

# Ensure X_daily, y_daily, features_daily are available from previous steps and not None
if ('X_daily' in locals() and X_daily.shape[0] > 0 and
    'y_daily' in locals() and y_daily.shape[0] > 0 and
    'features_daily' in locals() and len(features_daily) > 0):

    # Define a function to build the model with hyperparameters as arguments
    # We'll use the Simple_LSTM_V2 architecture as an example for tuning
    def build_tuned_lstm_model(lstm_units=64, dense_units=16, learning_rate=0.001, input_shape=None):
        model = Sequential([
            LSTM(lstm_units, return_sequences=True, input_shape=input_shape),
            LSTM(lstm_units // 2), # Reduce units in the second LSTM
            Dense(dense_units, activation='relu'),
            Dense(1, activation='sigmoid')
        ])
        optimizer = Adam(learning_rate=learning_rate)
        model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
        return model

    # Define a hyperparameter grid to search over
    param_grid = {
        'lstm_units': [32, 64],
        'dense_units': [8, 16],
        'learning_rate': [0.01, 0.001, 0.0001],
        'batch_size': [32, 64],
        'epochs': [15, 20] # Limit epochs for faster exploration
    }

    # Prepare data for tuning (using a fixed train/validation split for each combination)
    X_train_tune, X_val_tune, y_train_tune, y_val_tune = train_test_split(
        X_daily, y_daily, test_size=0.2, shuffle=False # Use shuffle=False for time series data
    )

    best_accuracy = 0
    best_params = {}
    tuning_results = []

    print("Starting Hyperparameter Tuning...")

    # Iterate through all combinations of hyperparameters
    for params in itertools.product(*param_grid.values()):
        current_params = dict(zip(param_grid.keys(), params))
        print(f"\nTraining with params: {current_params}")

        # Build and compile the model with current hyperparameters
        # Ensure input_shape is passed correctly
        model = build_tuned_lstm_model(
            lstm_units=current_params['lstm_units'],
            dense_units=current_params['dense_units'],
            learning_rate=current_params['learning_rate'],
            input_shape=(X_train_tune.shape[1], X_train_tune.shape[2])
        )

        # Train the model
        history = model.fit(
            X_train_tune, y_train_tune,
            epochs=current_params['epochs'],
            batch_size=current_params['batch_size'],
            validation_data=(X_val_tune, y_val_tune),
            verbose=0 # Keep training output clean
        )

        # Evaluate on the validation set
        loss, accuracy = model.evaluate(X_val_tune, y_val_tune, verbose=0)

        print(f"Validation Accuracy: {accuracy:.4f}")

        # Store results
        tuning_results.append({**current_params, 'val_accuracy': accuracy, 'val_loss': loss})

        # Check if this is the best model so far
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_params = current_params
            # Optionally save the best model here

    print("\nHyperparameter Tuning Complete.")
    print(f"Best Validation Accuracy: {best_accuracy:.4f}")
    print(f"Best Hyperparameters: {best_params}")

    # Optionally display all tuning results in a DataFrame
    tuning_results_df = pd.DataFrame(tuning_results)
    print("\nTuning Results Summary:")
    display(tuning_results_df.sort_values(by='val_accuracy', ascending=False))


else:
    print("Skipping hyperparameter tuning as data (X_daily, y_daily) or features (features_daily) are not available or empty.")
    print("Please ensure the data loading, feature engineering, and sequence creation steps were successful for daily data.")
```
